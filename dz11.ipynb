```python
stats = pd.read_csv('ratings.csv')
stats.head()
```

```python
def klassif(row):
    if row['rating'] >= 4.5: return 'высокий рейтинг'
    if row['rating'] <= 2.0: return 'низкий рейтинг'
    return 'средний рейтинг'
```

```python
stats['klass_']=stats.apply(klassif,axis=1)
```

```python
geo_data = {

'Центр': ['москва', 'тула', 'ярославль'],

'Северо-Запад': ['петербург', 'псков', 'мурманск'],

'Дальний Восток': ['владивосток', 'сахалин', 'хабаровск']
}
```

```python
stats = pd.read_csv('keywords.csv')
stats.head()
```

```python
def geo_reg(row):
    
    for i in geo_data:
        for g in geo_data[i]:
            if g[0:(len(g)-1)] in row['keyword']:
                #print(g[0:(len(g)-1)],row['keyword'])
                #print(i)
                return i
    return 'undefined'
```

```python
stats['region']=stats.apply(geo_reg, axis=1)
stats.head(3000)
```
